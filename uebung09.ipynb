## Aufgabe 1: Helizität und Chiralität

Erzeugen Sie den Spinor $u_+$ eines Myons mit positiver Helizität, das sich mit Energie $E$ in die Richtung ($\theta$, $\phi$) bewegt.

1. Wenden Sie die Chiralitätsprojektionsoperatoren $P_{R/L} = \frac{1}{2} \left( \mathbb{1} \pm \gamma^5 \right)$ auf $u_+$ an, um die Spinoren $u_R$ bzw. $u_L$ zu erzeugen.
2. Zeigen Sie, dass der Zustand $u_+$ keine definierte Chiralität hat.
3. In welchem Grenzfall hat $u_+$ eine definierte Chiralität? Welche Chiralität ist es?

Hinweis:
* Die $\gamma$-Matrizen `mgamma` beinhalten auch $\gamma^5$: `mgamma(5)`.

```python
import sys
from sympy import *
try:
    importlib.reload(sys.modules['spinors_heli'])  # ggf. Modul neu laden
except:
    pass
from spinors_heli import uH, uHbar, vH, vHbar
from sympy.physics.matrices import msigma,mgamma

E,m,Theta,Phi = symbols('E,m,Theta,Phi', positive=True)

PR = 1/2 * (eye(4) + mgamma(5))
PL = 1/2 * (eye(4) - mgamma(5))

u1  = uH(E,m,Theta,Phi,hel = '+')

uR, uL = PR * u1, PL * u1

print("u_+ \n")
pprint(u1)

print("\n u_R und u_L \n")
pprint(uR)
pprint(uL)

print("\n Für E >> m :  \n")

u1  = uH(E,0,Theta,Phi,hel = '+')

uR, uL = PR * u1, PL * u1

print("u_+ \n")
pprint(u1)

print("\n u_R und u_L \n")
pprint(uR)
pprint(uL)
```

## Aufgabe 2: Myonneutrino-Elektron-Streuung

Betrachten Sie die Streureaktion $\nu_\mu\,e^- \longrightarrow \mu^-\,\nu_e$ für Impulsüberträge, die klein im Vergleich zur Masse des W-Bosons sind. Die Elektronmasse soll vernachlässigt werden.

1. Wählen Sie ein geeignetes Koordinatensystem, und weisen Sie jedem Teilchen Impuls und Spinor zu. Die möglichen Spineinstellungen der Teilchen sollen durch Spinoren negativer bzw. positiver Helizität dargestellt werden.

2. Berechnen Sie die schwachen Viererströme und die Matrixelemente mit SymPy unter Verwendung der zuvor berechneten Spinoren.

3. Bestimmen Sie nun den differentiellen und den totalen Wirkungsquerschnitt. Wie ändert sich das Ergebnis, wenn sie auch die Myonmasse vernachlässigen können?

Hinweis: 
* Für die Teilaufgabe 1 sowie die Teile 2+3 gibt es jeweils einen Punkt.

```python
import sys
from sympy import *
try:
    importlib.reload(sys.modules['spinors_heli'])  # ggf. Modul neu laden
except:
    pass
from spinors_heli import uH, uHbar, vH, vHbar
from sympy.physics.matrices import msigma,mgamma

E,e,m,Theta,Phi,GF = symbols('E,e,m,Theta,Phi,GF', positive=True)

print('\n Nr. 1 \n')

u11  = uH(E,0,0,0,hel = '-')
v11b = uHbar(E,0,2*pi,0,hel = '-')
u12  = uH(E,0,0,0,hel = '+')
v12b = uHbar(E,0,2*pi,0,hel = '+')

u21  = uH(E,m,Theta,Phi,hel = '-')
v21b = uHbar(E,0,-Theta,-Phi,hel = '-')
u22  = uH(E,m,Theta,Phi,hel = '+')
v22b = uHbar(E,0,-Theta,-Phi,hel = '+')

#pprint(u1)
#pprint(u2)

j1,j2 = Matrix(4,1,[0,0,0,0]), Matrix(4,1,[0,0,0,0])

for i in range(4):
    j1[i] = v11b * mgamma(i, True) * (eye(4) - mgamma(5)) * u11
    j2[i] = v21b * mgamma(i, True)  * u21 + v22b * mgamma(i, True)  * u21 + v21b * mgamma(i, True)  * u22 + v22b * mgamma(i, True)  * u22

    
pprint(j1)
pprint(j2)
    
M = abs(GF/sqrt(2) * j1.dot(j2))**2

pprint(simplify(M))

```
