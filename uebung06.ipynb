## Aufgabe 1: Fouriertransformierte des Coulomb-Potentials

Zeigen Sie mit Hilfe von SymPy, dass die Fouriertransformierte des
Coulomb-Potentials durch die Formel 
\begin{equation*}
\int d^3 x\,\frac{\exp\left(-\imath\,\vec{q}\cdot\vec{x}\right)}{4\pi\left|\vec{x}\right|} = \frac{1}{\vec{q}^2}
\end{equation*}
gegeben ist.

Wechseln Sie dafür in ein Koordinatensystem mit geeigneter
Ausrichtung ($z$-Achse). Führen Sie an geeigneter Stelle der Rechnung
einen „konvergenzerzeugenden Faktor“ $\exp\left(-\lambda\,r\right)$ mit $\lambda \ge 0$
ein, und bestimmen Sie schließlich den Grenzfall $\lambda \rightarrow
0$.

Welches Ergebnis erhalten Sie ohne konvergenzerzeugenden Faktor?

```python
from sympy import *

init_printing(use_unicode=False, wrap_line=False)

x,r,z,q,qr,qz,c,phi,l = symbols('x,r,z,q,qr,qz,c,phi,l',positive=True)

qr=0

c = r*exp(-I*qz*z) * exp(-I*qr*r*cos(phi)) / (4*pi*sqrt(r**2+z**2))

pprint(c)
print('\n')

d = integrate(c,(phi,0,2*pi))

pprint(d)
print('\n')

e = integrate(d,(r,0,oo))

pprint(e)
print('\n')

e = -z*exp(-I*qz*z) * exp(-l*z) / 2

f = integrate(2*e,(z,0,oo))

pprint(f)
print('\n')

limit(f,l,0)
```
