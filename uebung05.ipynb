## Aufgabe 1: Rotationen und Lorentz-Boosts von Spinoren

Die Rotation eines Spinors um den Winkel $\theta$ um die $z$-Achse
kann durch folgende Matrix dargestellt werden:
\begin{equation*}
S^z_\text{Rot}\left(\theta\right) = 
\exp\left(-\frac{\theta}{2}\,\gamma^1\,\gamma^2\right) 
\end{equation*}
Die Matrix für Lorentz-Boosts entlang der $z$-Achse mit
$\omega=\tanh^{-1}\beta$ lautet:
\begin{equation*}
S^z_\text{Lor}\left(\omega\right) =
\exp\left(-\frac{\omega}{2}\,\gamma^0\,\gamma^3\right)
\end{equation*}

Implementieren Sie die Transformationsmatrizen $S^z_\text{Rot}$ und
$S^z_\text{Lor}$ mit Hilfe von SymPy als Python-Modul `lorentz`, um sie auch in
den folgenden Aufgaben nutzen zu können. 

1. Zeigen Sie, dass folgendes
gilt:
\begin{eqnarray*}
S^z_\text{Rot}\left(\theta\right) & = &
\mathbb{1}\,\cos\frac{\theta}{2} -
\gamma^1\,\gamma^2\,\sin\frac{\theta}{2},\\
S^z_\text{Lor}\left(\omega\right) & = &
\mathbb{1}\,\cosh\frac{\omega}{2} -
\gamma^0\,\gamma^3\,\sinh\frac{\omega}{2}
\end{eqnarray*}
2. Zeigen Sie auch, dass $S^z_\text{Rot}\left(-\theta\right)$ und
$S^z_\text{Lor}\left(-\omega\right)$ die inversen Matrizen zu $S^z_\text{Rot}\left(\theta\right)$ bzw. $S^z_\text{Lor}\left(\omega\right)$ sind.

## Aufgabe 2: Rotationen und $\gamma$-Matrizen

Überprüfen Sie für die Rotationsmatrix $S^z_\text{Rot} = S^z_\text{Rot}(\theta)$:
\begin{eqnarray*}
{S^z_\text{Rot}}^{-1}\,\gamma^0\,S^z_\text{Rot} & = & \gamma^0\\
{S^z_\text{Rot}}^{-1}\,\gamma^1\,S^z_\text{Rot} & = &
\gamma^1\,\cos\theta + \gamma^2\,\sin\theta\\
{S^z_\text{Rot}}^{-1}\,\gamma^2\,S^z_\text{Rot} & = &
-\gamma^1\,\sin\theta +\gamma^2\,\cos\theta\\
{S^z_\text{Rot}}^{-1}\,\gamma^3\,S^z_\text{Rot} & = & \gamma^3
\end{eqnarray*}

Tipp: Die Funktion `expand` kann auch die komplexe Exponentialfunktion in eine Summe von cos- und sin-Ausdrücken umwandeln, wenn man `expand(..., complex = True)` verwendet.

```python
from lorentz import Szrot

theta,omega = symbols('theta,omega')

lhs = simplify(Szrot(theta)**(-1) * mgamma(0) * Szrot(theta))
rhs = simplify(mgamma(0))

pprint([lhs,rhs])
print('\n',lhs.equals(rhs), '\n \n')


lhs = simplify(Szrot(theta)**(-1) * mgamma(1) * Szrot(theta))
rhs = simplify(mgamma(1) * cos(theta) + mgamma(2) * sin(theta))

pprint([lhs,rhs])
print('\n',lhs.equals(rhs), '\n \n')


lhs = simplify(Szrot(theta)**(-1) * mgamma(2) * Szrot(theta))
rhs = simplify((-mgamma(1) * sin(theta) + mgamma(2) * cos(theta)))

pprint([lhs,rhs])
print('\n',lhs.equals(rhs), '\n \n')


lhs = simplify(Szrot(theta)**(-1) * mgamma(3) * Szrot(theta))
rhs = simplify(mgamma(3))

pprint([lhs,rhs])
print('\n',lhs.equals(rhs), '\n \n')
```

## Aufgabe 3*: Lorentz-Boosts und $\gamma$-Matrizen (freiwillig)

Überprüfen Sie für die Lorentz-Boost-Matrix $S^z_\text{Lor} = S^z_\text{Lor}(\omega)$:
\begin{eqnarray*}
{S^z_\text{Lor}}^{-1}\,\gamma^0\,S^z_\text{Lor} & = &
\gamma^0\,\cosh\omega -\gamma^3\,\sinh\omega\\
{S^z_\text{Lor}}^{-1}\,\gamma^1\,S^z_\text{Lor} & = &
\gamma^1\\
{S^z_\text{Lor}}^{-1}\,\gamma^2\,S^z_\text{Lor} & = &
\gamma^2\\
{S^z_\text{Lor}}^{-1}\,\gamma^3\,S^z_\text{Lor} & = &
-\gamma^0\,\sinh\omega + \gamma^3\,\cosh\omega
\end{eqnarray*}

```python
from lorentz import Szlor

theta,omega = symbols('theta,omega')

lhs = simplify(Szlor(omega)**(-1) * mgamma(0) * Szlor(omega))
rhs = simplify(mgamma(0) * cosh(omega) - mgamma(3) * sinh(omega))

pprint([lhs,rhs])
print('\n',lhs.equals(rhs), '\n \n')


lhs = simplify(Szlor(omega)**(-1) * mgamma(1) * Szlor(omega))
rhs = simplify(mgamma(1))

pprint([lhs,rhs])
print('\n',lhs.equals(rhs), '\n \n')


lhs = simplify(Szlor(omega)**(-1) * mgamma(2) * Szlor(omega))
rhs = simplify(mgamma(2))

pprint([lhs,rhs])
print('\n',lhs.equals(rhs), '\n \n')


lhs = simplify(Szlor(omega)**(-1) * mgamma(3) * Szlor(omega))
rhs = simplify(-mgamma(0) * sinh(omega) + mgamma(3) * cosh(omega))

pprint([lhs,rhs])
print('\n',lhs.equals(rhs), '\n \n')
```
